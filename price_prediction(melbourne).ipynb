```python
fe=["Lattitude", "Longtitude", "Distance", "Regionname","Type" ,"Rooms", "Bedroom2", "Bathroom"
    , "Landsize","BuildingArea","YearBuilt", "Car", "CouncilArea"]
target="Price"
```

```python
df.info()
```

```python
x=df[fe]
y=df[target]
```

```python
cont1=x.select_dtypes(include=["int64","float64"]).columns
cat1=x.select_dtypes(include=["object"]).columns
cat1
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
con=Pipeline(steps=[("simcont",SimpleImputer(strategy="median")),("stand",StandardScaler())])
```

```python
cat=Pipeline(steps=[("simcat",SimpleImputer(strategy="most_frequent")),("ohe",OneHotEncoder())])
```

```python
proc=ColumnTransformer(transformers=[("cont",con,cont1),("cat",cat,cat1)])
```

```python
model=Pipeline(steps=[("processed",proc),("model1",RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1))])
```

```python
model
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.3,random_state=23)
```

```python
model.fit(xTrain,yTrain)
```

```python
ypred=model.predict(xTest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(yTest, ypred)
mse = mean_squared_error(yTest, ypred)
r2 = r2_score(yTest, ypred)


print(f"Mean Absolute Error (MAE): {mae:,.2f}")
print(f"Mean Squared Error (MSE): {mse:,.2f}")
print(f"R-squared (R2 Score): {r2:.4f}")
```
